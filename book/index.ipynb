# RF Signal Classification

![Project Overview](figures/project_overview.png)

## Project Overview
This project builds an end-to-end pipeline to classify RF (Radio Frequency) 
signals by their modulation type using the RadioML 2016.10a dataset.

## What We Cover
- Signal processing (FFT, PSD, Constellation diagrams)
- Feature engineering from raw IQ samples
- SQL database design and querying
- Machine learning classification (XGBoost)
- Model explainability with SHAP

## Dataset
- **Source:** RadioML 2016.10a (DeepSig)
- **Signals:** 11 modulation types
- **SNR range:** -20dB to +18dB
- **Total samples:** 220,000 IQ signals

## Tech Stack
- Python 3.12
- NumPy / Pandas
- SQLite / SQLAlchemy
- Scikit-learn / XGBoost
- Plotly / Seaborn / Bokeh
- Streamlit
- Quarto (this book)

## Project Structure

```
rf-signal-classifier/
│
├── data/
│   ├── raw/                    # Original RadioML .hdf5 dataset - never modified
│   └── processed/              # Cleaned CSVs with extracted features
│
├── book/                       # Quarto book - published as GitHub Pages
│   ├── _quarto.yml             # Book configuration (theme, chapters, format)
│   ├── index.ipynb             # Project overview (this page)
│   └── chapters/               # One notebook per chapter
│       ├── 01_introduction.ipynb       # Background on RF signals and modulation
│       ├── 02_data_exploration.ipynb   # Loading and exploring IQ samples
│       ├── 03_signal_processing.ipynb  # FFT, PSD, constellation diagrams
│       ├── 04_feature_engineering.ipynb # Statistical and spectral features
│       ├── 05_modeling.ipynb           # XGBoost classifier training
│       └── 06_results.ipynb            # Evaluation, SHAP, accuracy vs SNR
│
├── src/                        # Python source modules (reusable functions)
│   ├── __init__.py             # Makes src a Python package
│   ├── data_loader.py          # Parses HDF5, loads IQ samples into memory
│   ├── signal_processing.py    # FFT, PSD, spectrogram computations
│   ├── features.py             # Extracts statistical and spectral features
│   ├── database.py             # SQLite schema, inserts, and queries
│   └── model.py                # Trains, evaluates and saves classifier
│
├── sql/
│   ├── schema.sql              # CREATE TABLE statements for SQLite DB
│   └── queries.sql             # Analysis queries (GROUP BY, aggregations)
│
├── reports/
│   └── figures/                # All saved plots (PNG) for the book and README
│       ├── constellation/      # I vs Q scatter plots per modulation type
│       ├── fft/                # Power spectrum plots
│       ├── evaluation/         # Confusion matrix, accuracy vs SNR curves
│       └── shap/               # SHAP feature importance plots
│
├── app/                        # Streamlit interactive web application
│   └── main.py                 # Upload signal → classify → visualize results
│
├── tests/                      # Unit tests with pytest
│   ├── test_data_loader.py     # Tests for HDF5 parsing functions
│   ├── test_signal_processing.py # Tests for FFT and PSD functions
│   └── test_features.py        # Tests for feature extraction functions
│
├── .github/
│   └── workflows/
│       └── ci.yml              # GitHub Actions - runs tests on every push
│
├── config.yaml                 # Central config (paths, model params, DB settings)
├── .env                        # Secret variables - never pushed to GitHub
├── .python-version             # Locks Python version via pyenv (3.12.0)
├── .gitignore                  # Files excluded from Git
├── requirements.txt            # All Python dependencies with exact versions
└── README.md                   # Project homepage on GitHub
```
